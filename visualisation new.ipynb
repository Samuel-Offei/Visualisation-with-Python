```python
# Create a box plot
plt.boxplot(data['Price'])

# Add labels and title
plt.xlabel('Price')
plt.ylabel('Values')
plt.title('Box Plot Example')

# Show the plot
plt.show()
```

```python
# Create a box plot
plt.boxplot(data['Rating'])

# Add labels and title
plt.xlabel('Rating')
plt.ylabel('Values')
plt.title('Box Plot Example')

# Show the plot
plt.show()
```

```python
# Create a box plot
plt.boxplot(data['Reviews'])

# Add labels and title
plt.xlabel('Reviews')
plt.ylabel('Values')
plt.title('Box Plot Example')

# Show the plot
plt.show()
```

```python
# Create a box plot
plt.boxplot(data['Number_Of_Pages'])

# Add labels and title
plt.xlabel('Reviews')
plt.ylabel('Values')
plt.title('Box Plot Example')

# Show the plot
plt.show()
```

```python
# compute correlation
corr_matrix = data[['Price','Rating', 'Reviews', 'Number_Of_Pages']].corr()

corr_matrix
```

```python
# plot heatmap
# 'annot=True' returns the correlation values 
sns.heatmap(data[['Price','Rating', 'Reviews', 'Number_Of_Pages']].corr(), annot = True)

# display the plot
plt.show()
```
